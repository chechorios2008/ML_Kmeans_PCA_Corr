<h3>1. Realice un ejercicio de agrupación utilizando el algoritmo K-means para las variables longitude y latitude, en los siguientes escenarios de 2, 3 y 4 grupos. </h3>

```python
K = 2

kmeans = KMeans(n_clusters=K, random_state=42)
kmeans.fit(df[['longitude', 'latitude']])
df['cluster'] = kmeans.labels_

plt.figure(figsize=(7, 4))
plt.scatter(x=df['longitude'], y=df['latitude'], c=df['cluster'], cmap='viridis')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.title('Agrupación K-means (K = 2)')
plt.show()
```

```python
K = 3

kmeans = KMeans(n_clusters=K, random_state=42)

kmeans.fit(df[['longitude', 'latitude']])

df['cluster'] = kmeans.labels_

plt.figure(figsize=(7, 4))
plt.scatter(x=df['longitude'], y=df['latitude'], c=df['cluster'], cmap='viridis')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.title('Agrupación K-means (K = 3)')
plt.show()
```

```python
K = 4
kmeans = KMeans(n_clusters=K, random_state=42)
kmeans.fit(df[['longitude', 'latitude']])
df['cluster'] = kmeans.labels_

plt.figure(figsize=(7, 4))
plt.scatter(x=df['longitude'], y=df['latitude'], c=df['cluster'], cmap='viridis')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.title('Agrupación K-means (K = 4)')
plt.show()
```

<h3>2. ¿De acuerdo con los resultados del punto anterior con que cantidad de grupos te quedarías y por qué? </h3>

<h2> K=3 </h2>
<p> Se hace uso de la grafica del codo evaluando los posibles valores para el hiperparametro K, La gráfica del codo es una herramienta útil para identificar el valor de K que podría ser óptimo, pero no es una regla absoluta. En el ejercicio se aprecia que K=3 es el posible valor que más describiría los datos analisados </p>

```python
range_values = list(range(2, 5))

# Lista para almacenar los valores de SSE
sse_values = []

for k in range_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df[['longitude', 'latitude']])
    sse_values.append(kmeans.inertia_)  # SSE para cada K

diff_sse_values = [sse_values[i+1] - sse_values[i] for i in range(len(sse_values)-1)]

plt.figure(figsize=(7, 4))
plt.plot(range_values, sse_values)
plt.xlabel('Número de clusters (K)')
plt.ylabel('Suma del error cuadrático dentro de los clusters (SSE)')
plt.title('Método del codo para selección de K')
plt.grid(True)
plt.show()
```

<h3>4. Realice la reducción del número de variables conservando el máximo de información posible. Este ejercicio hágalo mediante el algoritmo de análisis de componentes principales. </h3>

```python
scaler = StandardScaler()
df_escalado = scaler.fit_transform(df)
```

```python
# Escalado del set de datos con las variables a trabajar
df_escalado = pd.DataFrame(df_escalado, columns=df.columns)
df_escalado.head()
```

<h3> PCA </h3>

```python
# Aplicando PCA a los datos escalados.
pca = PCA()
pca.fit(df_escalado)
pca_df_escalado = pca.transform(df_escalado)
pca_df_escalado
```

```python
# Analisis del indice de varianza: El objetivo es hacer una suma acumulada de cada valor para determinar la cantidad de valores.
var = pca.explained_variance_ratio_
print(var)
```

```python
cum_var = np.cumsum(np.round(var, decimals=4)*100)
plt.figure(figsize=(7, 4))
plt.plot(cum_var, 'b-o')
plt.show()
```

```python
pca_df = pd.DataFrame(pca_df_escalado)
pca_df.drop([5,6,7,8, 9],axis=1, inplace=True)
pca_df.head(10)
```

<h3>5. Considera usted que para el conjunto de variables presentado es una buena decisión realizar una reducción de variables, las componentes encontradas logran hacer una buena representación de las variables originales. </h3>

<h3> La aplicación de PCA al conjunto de variables presentadas es una decisión acertada.  PCA representan el 90% en 4 variables</h3>
<p>
<li> Reducción significativa de la dimensionalidad: Se ha logrado una reducción considerable de la cantidad de variables, pasando de 9 a solo 4 componentes principales. Esto implica una simplificación del análisis y una menor complejidad en la interpretación de los resultados. </li>
<li> Conservación de información relevante: A pesar de la reducción de variables, las 4 componentes principales capturadas por PCA representan el 90% de la varianza total de los datos originales. Esto indica que la información esencial del conjunto de datos se conserva en gran medida en las componentes reducidas.  </li>
<li> Interpretación más clara de las relaciones: Al trabajar con un menor número de variables, se facilita la identificación de patrones y relaciones entre las variables originales. Las componentes principales representan combinaciones lineales de las variables originales, lo que permite comprender mejor cómo se correlacionan entre sí. </li>
<li> Eficiencia computacional:La reducción de la dimensionalidad mediante PCA también conlleva ventajas computacionales.  </li>
<li> Análisis del índice de varianza: Las primeras 4 componentes principales capturan la mayor parte de la varianza, mientras que las componentes restantes aportan una contribución menor. </li>
